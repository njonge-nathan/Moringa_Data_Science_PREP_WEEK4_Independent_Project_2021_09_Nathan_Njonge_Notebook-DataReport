```
#check to see if certain columns have similar values and hence drop one.
# From the description above the cars and Bluecar counter seem to have similar properies, check to see if they are actually similar value
Check= Autolib.Cars==Autolib['Bluecar counter']
Check.describe()
```

```
# From the questions we know we need the postal code column.
# We will drop the ID and public name column since we don't need them
Bluecars= Autolib.drop(['Cars','Utilib counter','Utilib 1.4 counter','Displayed comment','ID','Kind','Geo point','Public name','Subscription status','Scheduled at'], axis=1)
Bluecars.head()
```

```
# For ease in analysis we combine the year, month, day, hour and minute columns to form one column named Date-Time
Bluecars.year = Bluecars.year.apply(str)
Bluecars.month = Bluecars.month.apply(str)
Bluecars.day = Bluecars.day.apply(str)

Bluecars.hour = Bluecars.hour.apply(str)
Bluecars.minute = Bluecars.minute.apply(str)

Bluecars.info()
```

```
Date = Bluecars['year'] + Bluecars['month'] + Bluecars['day']
Test = pd.to_datetime(Date, format='%Y%m%d', errors='ignore')
print(Test.head())
Time = Bluecars.hour + Bluecars.minute
Time_test= pd.to_datetime(Time, format= '%H%M')- pd.to_datetime(Time, format='%H%M').dt.normalize()
print(Time_test)
```

```
Bluecars['Date']= Test
Bluecars
```

```
Bluecars.drop(['year', 'month', 'day', 'minute'], axis=1, inplace= True)
Bluecars.head()
```

```
Bluecars.hour = Bluecars.hour.apply(int)
Bluecars.info()
```

#DATA ANALYSIS

```
# Our main city of focus is Paris
Bluecars = Bluecars[Bluecars.City == 'Paris']
Bluecars
```

```
# sort in ascending order based on days
Bluecars= Bluecars.sort_values(by=['Address','Date', 'hour'])
Bluecars
```

```
# We need to see how many cars were taken and returned to the same address.
Active_Addresses = Bluecars.copy(deep=True)
Active_Addresses = Active_Addresses[Active_Addresses['Address'] == Active_Addresses['Address'].shift(1)]
Active_Addresses.loc[Active_Addresses.Address == Active_Addresses.Address.shift(-1), 'Pickup'] = Active_Addresses['Bluecar counter'] - Active_Addresses['Bluecar counter'].shift(-1)
Active_Addresses.head()
```

```
Active_Addresses.Pickup.fillna(0, inplace= True)
Active_Addresses.Pickup.value_counts()
```

```
Active_Addresses.Pickup = Active_Addresses.Pickup.replace([-1,-2,-3,-4,-5,-6,-7], 0)
Active_Addresses.Pickup.value_counts()
```

#most popular hours

```
  Engaged_hours = Active_Addresses.copy(deep= True)
Weekday_hours = Engaged_hours[((Engaged_hours.hour >= 8) & (Engaged_hours.hour <= 18)) & ((Engaged_hours.Date != '2018-04-01') & (Engaged_hours.Date != '2018-04-07') & (Engaged_hours.Date != '2018-04-08'))]
Weekday_hours.groupby('hour')['Pickup'].sum().sort_values(ascending=False)
```
